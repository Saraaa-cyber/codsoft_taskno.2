```python
from tkinter import*
root=Tk()
root.resizable(0,0)
root.geometry("320x400")
root.title("Calculator")
root.config(bg='azure')

y=StringVar()
ent=Entry(root,textvariable=y,borderwidth=6,font=("aerial",18),justify="right")
ent.place(width=320,height=45)

def points(p):
    x=y.get()
    if  len(x)==1 and  x[0]=="0":
        y.set(p)
    else:
        y.set(x+p)
def op(o):
    x=y.get()
    y.set(x+o)

def results():
    x=y.get()
    if ("+" in x):
        z=list(map(float,x.split("+")))
        sum=0
        for i in z:
            sum+=i
        y.set(sum)
    elif ("-" in x):
        z=list(map(float,x.split("-")))
        y.set(z[0]-z[1])
    elif ("*" in x):
        z=list(map(float,x.split("*")))
        y.set(z[0]*z[1])
    elif ("/" in x):
        z=list(map(float,x.split("/")))
        y.set(z[0]/z[1])
    
    elif ("√" in x):
        z=x.split("√")
        y.set(float(z[1])**0.5)
    elif ("!" in x):
        z = x.split("!")

        n=int(z[0])
            
        if n < 0:
            pass
        elif n == 0 or n == 1:
            f=1
        else:
            
            f = 1
            for i in range(2, int(z[0]) + 1):
                f *= i
        y.set(f)
        
        
    elif ("^" in x):
        z=x.split("^")
        base= float(z[0])
        expo= float(z[1])
        k= base**expo
        y.set(k)
    else:
        y.set("syntax error")

def clear():
    y.set("0")

C=Button(root,text="C",padx=15,pady=15,command=clear)
b7=Button(root,text="7",padx=15,pady=15,command=lambda:points("7"))
b4=Button(root,text="4",padx=15,pady=15,command=lambda:points("4"))
b1=Button(root,text="1",padx=15,pady=15,command=lambda:points("1"))
f=Button(root,text="!",padx=15,pady=15,command=lambda:op("!"))
C.place(x=20,y=53)
b7.place(x=20,y=123)
b4.place(x=20,y=193)
b1.place(x=20,y=263)
f.place(x=20,y=333)

sr=Button(root,text="√",padx=15,pady=15,command=lambda:op("√"))
b8=Button(root,text="8",padx=15,pady=15,command=lambda:points("8"))
b5=Button(root,text="5",padx=15,pady=15,command=lambda:points("5"))
b2=Button(root,text="2",padx=15,pady=15,command=lambda:points("2"))
b0=Button(root,text="0",padx=15,pady=15,command=lambda:points("0"))

sr.place(x=100,y=53)
b8.place(x=100,y=123)
b5.place(x=100,y=193)
b2.place(x=100,y=263)
b0.place(x=100,y=333)

di=Button(root,text="/",padx=15,pady=15,command=lambda:op("/"))
b9=Button(root,text="9",padx=15,pady=15,command=lambda:points("9"))
b6=Button(root,text="6",padx=15,pady=15,command=lambda:points("6"))
b3=Button(root,text="3",padx=15,pady=15,command=lambda:points("3"))
do=Button(root,text=".",padx=15,pady=15,command=lambda:points("."))

di.place(x=180,y=53)
b9.place(x=180,y=123)
b6.place(x=180,y=193)
b3.place(x=180,y=263)
do.place(x=180,y=333)

ex=Button(root,text="^",padx=15,pady=15,command=lambda:op("^"))
mul=Button(root,text="*",padx=15,pady=15,command=lambda:op("*"))
sub=Button(root,text="-",padx=15,pady=15,command=lambda:op("-"))
ad=Button(root,text="+",padx=15,pady=15,command=lambda:op("+"))
eq=Button(root,text="=",padx=15,pady=15,command=results)

ex.place(x=260,y=53)
mul.place(x=260,y=123)
sub.place(x=260,y=193)
ad.place(x=260,y=263)
eq.place(x=260,y=333)




root.mainloop()
```
